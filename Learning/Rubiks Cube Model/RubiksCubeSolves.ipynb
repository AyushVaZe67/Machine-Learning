```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('solves.csv')
```

```python
df
```

```python
df = df.drop(columns=['#'])
```

```python
df.info()
```

```python
df = df.replace("DNF", np.nan)
```

```python
df.sample(10)
```

```python
df[["Solve 1", "Solve 2", "Solve 3", "Solve 4", "Solve 5"]] = df[["Solve 1", "Solve 2", "Solve 3", "Solve 4", "Solve 5"]].astype(float)
```

```python
df.info()
```

```python
df[['Fastest Solve']] = df[['Fastest Solve']].astype(float)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot a heatmap for missing values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Count missing values in each column
missing_values = df.isnull().sum()

# Plot
missing_values.plot(kind='bar', figsize=(8, 6), color='orange')
plt.title("Missing Values in Each Column")
plt.ylabel("Number of Missing Values")
plt.xlabel("Columns")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numerical features
numeric_columns = ["Solve 1", "Solve 2", "Solve 3", "Solve 4", "Solve 5", "Fastest Solve"]

# Pair plot
sns.pairplot(df, vars=numeric_columns, hue="Country", palette="viridis", corner=True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot violin plot for solve times
solve_columns = ["Solve 1", "Solve 2", "Solve 3", "Solve 4", "Solve 5"]
plt.figure(figsize=(10, 6))
sns.violinplot(data=df[solve_columns], palette="muted", scale="width")
plt.title("Violin Plot of Solve Times")
plt.ylabel("Solve Time (seconds)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.api.types import is_numeric_dtype

# Prepare data for ridge plot
melted_df = df.melt(value_vars=["Solve 1", "Solve 2", "Solve 3", "Solve 4", "Solve 5"], var_name="Solve", value_name="Time")

# Plot ridge plot
plt.figure(figsize=(10, 8))
sns.violinplot(x="Time", y="Solve", data=melted_df, palette="viridis", bw=0.2, cut=0)
plt.title("Ridge Plot of Solve Times")
plt.xlabel("Solve Time (seconds)")
plt.ylabel("Solve Attempt")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numeric columns
numeric_columns = ["Solve 1", "Solve 2", "Solve 3", "Solve 4", "Solve 5", "Fastest Solve"]

# Pair plot with KDE
sns.pairplot(df[numeric_columns], diag_kind="kde", palette="coolwarm")
plt.suptitle("Pairwise Distributions with KDE", y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")

# Scatter
ax.scatter(df["Solve 1"], df["Solve 2"], df["Solve 3"], c=df["Fastest Solve"], cmap="viridis", s=50)

# Labels
ax.set_xlabel("Solve 1")
ax.set_ylabel("Solve 2")
ax.set_zlabel("Solve 3")
plt.title("3D Scatter Plot")
plt.show()
```
