```python
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])
plt.title('Rohit Sharma VS Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')
plt.plot(batsman['index'],batsman['V Kohli'],color='#FC00D6')
plt.title('Rohit Sharma VS Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='solid',linewidth=5)
plt.plot(batsman['index'],batsman['V Kohli'],color='#FC00D6',linestyle='dashed')
plt.title('Rohit Sharma VS Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='solid',linewidth=1,marker='+',markersize=10,label='Rohit')
plt.plot(batsman['index'],batsman['V Kohli'],color='#FC00D6',linestyle='dashed',marker='o',markersize=5,label='Virat')
plt.title('Rohit Sharma VS Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs')

plt.legend()
```
