```python
df = pd.read_csv('batsman_season_record.csv')
```

```python
df
```

```python
plt.bar(df['batsman'],df['2015'],width=0.95)
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2, label='2015',color='purple')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, label='2016',color='pink')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2, label='2017',color='magenta')
plt.xticks(np.arange(df.shape[0]),df['batsman'])

plt.legend()
plt.show()
```
