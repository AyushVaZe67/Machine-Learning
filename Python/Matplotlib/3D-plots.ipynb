```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
x
```

```python
y
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
xx.shape
```

```python
yy.shape
```

```python
z = xx**2 + yy**2
```

```python
z.shape
```

```python
fig = plt.figure(figsize=(8,10))
ax = plt.subplot(projection='3d')
var = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(var)
```

```python
z = np.sin(xx**2+yy**2)
fig = plt.figure(figsize=(8,10))
ax = plt.subplot(projection='3d')
var = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(var)
```

```python
fig = plt.figure(figsize=(8,10))
ax = plt.subplot()
var = ax.contour(xx,yy,xx**2 + yy**2,cmap='viridis')
fig.colorbar(var)
```

```python
fig = plt.figure(figsize=(8,10))
ax = plt.subplot()
var = ax.contourf(xx,yy,xx**2 + yy**2,cmap='viridis')
fig.colorbar(var)
```

```python
z = np.sin(xx**2+yy**2)
fig = plt.figure(figsize=(8,10))
ax = plt.subplot()
var = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(var)
```
