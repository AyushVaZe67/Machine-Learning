```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
```

```python
tips = pd.read_csv('tips.csv')
```

```python
tips.head()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip', style='time', hue='sex', size='size')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex')
```

```python
gap = px.data.gapminder()
```

```python
gap
```

```python
temp_df = gap[gap['country']=='India']
```

```python
temp_df
```

```python
sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
temp_df = gap[gap['country'].isin(['India','Pakistan','China'])]
```

```python
temp_df
```

```python
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country')
```

```python
temp_df = gap[gap['country'].isin(['India','Norway','Italy'])]
```

```python
temp_df
```

```python
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='continent', style='country')
```
